# Pré-Processamento

```python
crunch.columns = ['id', 'titulo', 'tipo', 'descricao', 'ano_lanc', 'age_certification', 'runtime', 'generos', 'paises_producao', 'seasons', 'imdb_id', 'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']
```

```python
#Quantidade de animes/filmes de acordo com a categoria
generos = list()

for position in crunch['generos']:
    for i in position.split("'"):
        if i not in [']', ', ', '[', '[]']:
            if i not in generos:
                generos.append(i)
print(generos)
```

```python
# Removendo colunas sem uso
crunchyroll = crunch.drop(['id', 'titulo', 'tipo', 'ano_lanc', 'age_certification', 'runtime', 'paises_producao', 'seasons', 'imdb_id', 'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score'], axis = 1)
crunchyroll
```

```python
# Drop em nulls e em generos com lista vazia
crunchyroll.dropna(subset=['descricao'], inplace=True)
crunchyroll.query("generos != ['[]']", inplace = True)
crunchyroll
```

# Hipótese

```python
# Função para geração de colunas bit
def genres_to_bit(lista_generos, table, nome_genero):
    for genero in lista_generos:
        column = list()
        for linha_genero in table[nome_genero]:
            if genero in linha_genero:
                column.append(1)
            else:
                column.append(0)
        table[genero] = column
        
genres_to_bit(generos, crunchyroll, "generos")
crunchyroll.tail()
```
