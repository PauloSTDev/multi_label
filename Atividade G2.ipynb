```python
import pandas as pd
table = pd.read_csv('crunchyroll.csv')
table.columns = ['id', 'titulo', 'tipo', 'descricao', 'ano_lanc', 'age_certification', 'runtime', 'generos', 'paises_producao', 'seasons', 'imdb_id', 'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']
crunchyroll = table.drop(['id', 'tipo', 'ano_lanc', 'age_certification', 'runtime', 'paises_producao', 'seasons', 'imdb_id', 'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score'], axis = 1)
crunchyroll
```
